```python
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
movies = pd.read_excel('data/movies_df_without_embeddings.xlsx')
```

```python
movies
```

```python
embeddings = []
num_dimensions = 768
for index, row in tqdm(movies.iterrows()):
    overview = row["overview"]
    #print(overview)
    if isinstance(overview, str):
        embedding = list(model.encode(overview))
    else:    
        embedding = [0.0 for _ in range(num_dimensions)]
        
    embeddings.append(embedding)
movies["overview_embeddings"] = embeddings    
```

```python
movies = movies.set_index('movieId')
```
